```python
import requests

URL="https://edition.cnn.com/world/live-news/coronavirus-pandemic-vaccine-updates-03-11-21/index.html"
```

```python
page=requests.get(URL)
print(page.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
# Get main content <div>
news_content=soup.find(id='posts-and-button')

news_articles=news_content.find_all('article')
```

```python
news_list1= news_articles[1]
```

```python
news_list1.contents
```

```python
# title, date and author is in 0
news_list1.contents[2]
# article is in 1
#news_list.contents[1]
```

```python
head = news_list1.contents[0].text.strip()
#title date and author should be separated into 3 strings

#separate date and time from the title
date_time = head.split("Key")[0]
#separate date and time
date = date_time.split(", ",1)[1]

#separate title and author
author_date= head.split("From")
author = author_date[1]
title = author_date[0].split("2021")[1]
article = news_list1.contents[1].text.strip()

#verify if title date and author is separated properly
print(f'{title}')
print(f'{date}')
print(f'{author}')
```

## Save to Json 

```python
import json
article1_json = []

head = news_list1.contents[0].text.strip()
#title date and author should be separated into 3 strings

#separate date and time from the title
date_time = head.split("Key")[0]
#separate date and time
date = date_time.split(", ",1)[1]

#separate title and author
author_date= head.split("From")
author = author_date[1]
title = author_date[0].split("2021")[1]
article = news_list1.contents[1].text.strip()

article1_json.append({
    "title": title,
    "author": author,
    "date": date,
    "article": article
})

with open('article1.json', 'w') as json_file:
    json.dump(article1_json, json_file)

article1_json
```
