```python
import datetime
import json
cnn_json = []
# theres 613 articles, and it starts at 4, the limit is set to 24 to limit the number of articles
for i in range(4, 24, 1):
    cnn_article = cnn_paper.articles[i]
    cnn_article.download()
    cnn_article.parse()
    
    source = cnn_article.url
    title = cnn_article.title
    article_body = cnn_article.text
    author = cnn_article.authors

    date = cnn_article.publish_date

    
    if (date == None):
        cnn_json.append({
            "author": author,
            "title": title,
            "source": source,
            "article_body": article_body
        })
    else:
        datestring = date.strftime('%Y/%m/%d')
        cnn_json.append({
            "author": author,
            "title": title,
            "date": datestring,
            "source": source,
            "article_body": article_body
        })
        
with open('cnn_articles.json', 'w') as json_file:
    json.dump(cnn_json, json_file)

cnn_json
```
