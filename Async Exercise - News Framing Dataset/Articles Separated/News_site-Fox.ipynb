```python
import datetime
import json
fox_json = []
# theres 613 articles, and it starts at 4, the limit is set to 24 to limit the number of articles
for i in range(13, 34, 1):
    fox_article = fox_paper.articles[i]
    fox_article.download()
    fox_article.parse()
    
    source = fox_article.url
    title = fox_article.title
    article_body = fox_article.text
    author = fox_article.authors

    date = fox_article.publish_date

    
    if (date == None):
        fox_json.append({
            "author": author,
            "title": title,
            "source": source,
            "article_body": article_body
        })
    else:
        datestring = date.strftime('%Y/%m/%d')
        fox_json.append({
            "author": author,
            "title": title,
            "date": datestring,
            "source": source,
            "article_body": article_body
        })
        
with open('fox_articles.json', 'w') as json_file:
    json.dump(fox_json, json_file)

fox_json
```
