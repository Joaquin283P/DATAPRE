```python
import datetime
import json
nyt_json = []
# theres 613 articles, and it starts at 4, the limit is set to 24 to limit the number of articles
for i in range(0,20, 1):
    nyt_article = nyt_paper.articles[i]
    nyt_article.download()
    nyt_article.parse()
    
    source = nyt_article.url
    title = nyt_article.title
    article_body = nyt_article.text
    author = nyt_article.authors

    date = nyt_article.publish_date

    
    if (date == None):
        nyt_json.append({
            "author": author,
            "title": title,
            "source": source,
            "article_body": article_body
        })
    else:
        datestring = date.strftime('%Y/%m/%d')
        nyt_json.append({
            "author": author,
            "title": title,
            "date": datestring,
            "source": source,
            "article_body": article_body
        })
        
with open('nyt_articles.json', 'w') as json_file:
    json.dump(nyt_json, json_file)

nyt_json
```
